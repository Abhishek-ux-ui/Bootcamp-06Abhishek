```python
import cv2
from google.colab import files
import matplotlib.pyplot as plt
import numpy as np
```

```python
uploaded = files.upload()
```

```python
image_filename = next(iter(uploaded))
image = cv2.imread(image_filename)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
height, width, channels = image.shape
print(f"Image dimensions: Height = {height}, Width = {width}, Channels = {channels}")
```

```python
resized_image = cv2.resize(image, (400, 256))
```

```python
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.title('Resized Image (400x256)')
plt.axis('off')
plt.show()
```

```python
flipped_horizontally = cv2.flip(resized_image, 1)
flipped_vertically = cv2.flip(resized_image, 0)
```

```python
plt.imshow(cv2.cvtColor(flipped_horizontally, cv2.COLOR_BGR2RGB))
plt.title('Flipped Horizontally')
plt.axis('off')
plt.show()
```

```python
plt.imshow(cv2.cvtColor(flipped_vertically, cv2.COLOR_BGR2RGB))
plt.title('Flipped Vertically')
plt.axis('off')
plt.show()
```
