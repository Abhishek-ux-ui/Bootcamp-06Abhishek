```python
import nltk
from nltk.corpus import stopwords
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
nltk.download('stopwords')
```

```python
text = """
Nature is the physical world around us. It includes everything from the smallest microbe to the largest mountain range,
and encompasses a variety of ecosystems, climates, and habitats. Nature has always been an integral part of human life.
We rely on natural resources like water, air, and soil for survival. The natural world also provides us with inspiration
and opportunities for leisure, exploration, and scientific discovery. However, human activities have significantly altered
nature, leading to issues like climate change, deforestation, and pollution. It is important for us to take care of nature
in order to preserve the balance of life on Earth.
"""
```

```python
stop_words = set(stopwords.words('english'))
```

```python
wordcloud = WordCloud(stopwords=stop_words, background_color='white').generate(text)
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
