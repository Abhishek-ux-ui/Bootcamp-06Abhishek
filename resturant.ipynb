```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
```

```python
smoker_counts = tips['smoker'].value_counts()
plt.figure(figsize=(7, 7))
plt.pie(smoker_counts, labels=smoker_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel')[0:2])
plt.title('Percentage of Smokers vs Non-Smokers')
plt.show()
```
